Creating New, Renaming & Dropping Columns
Apply method, Sort Index, Sort Values

```python
import pandas as pd
data=pd.read_csv(r'D:\Data Analytics\UnifiedMentor\students.csv')
data
```

Create new Columns

```python
data['City']=['Delhi','Gurgaon','Gurgaon','jhansi','jhansi']
data
```

Renaming Columns

```python
data.columns=['Student','Age','Univesity','Location']
data
```

Dropping Columns

```python
data.drop(['Age'],axis=1) # temporary delete
data
```

```python
data.drop(['Age'],axis=1,inplace=True) # permanently delete
data
```

Apply Method

```python
# Creating Marks column
data['Marks']=[40,33,34,45,50]
data
```

```python
#Apply method
data['Marks']=data['Marks'].apply(lambda x:x/2)
data
```

Renaming Columns

```python
data.columns=['Name','University','City','Marks']
data
```

Sort Index

```python
data.index=[10,25,34,34,4]
data
```

```python
data.sort_index() # default ascending true  temoporary
res=data.sort_index()

res
```
